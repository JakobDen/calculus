```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Latex
from sympy import *
```

$p_{n+1} = 2^n\sqrt{2(1 - \sqrt{1 - (p_{n}/2^n)^2})}$

$p_3 = 3\sqrt{3}$

```python
p = [0, 0, 0, 3 * np.sqrt(3)]
for i in range(3, 61):
    p_next = np.power(2, i) * np.sqrt(2 * (1 - np.sqrt(1 - np.power(p[i] * np.float_power(2, -i), 2))))
    p.append(p_next)
plt.plot(np.arange(62), p)
plt.plot(np.arange(62), np.ones(62) * np.pi)
```

Как видно, синий график далеко не сошелся к $\pi$.
